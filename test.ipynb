```python
# Pivot to get counts per party
vote_counts = vote_data.pivot_table(
    index='party', 
    columns='option', 
    aggfunc='size', 
    fill_value=0
)

# Convert counts to ratios per party
vote_ratios = vote_counts.div(vote_counts.sum(axis=1), axis=0).reset_index()

print(vote_ratios)

source = ColumnDataSource(vote_ratios)

stackers = vote_ratios.columns.tolist()
stackers.remove('party')

colors = Category10[len(stackers)]

p = figure(
    y_range=vote_ratios['party'],
    sizing_mode='stretch_width',
    toolbar_location=None,
    title="Voting Results Ratio by Party"
)

p.hbar_stack(stackers, y='party', height=0.8, color=colors, source=source, legend_label=stackers)

p.y_range.range_padding = 0.1
p.ygrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_right"
p.legend.orientation = "horizontal"

show(p)
```
